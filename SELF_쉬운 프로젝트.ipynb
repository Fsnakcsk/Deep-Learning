## 라이브러리 Load

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
```

## data load

```python
data = pd.read_csv('gpascore.csv')
```

```python
data.isna().sum()
```

```python
data = data.dropna()
```

```python
# data.fillna()
```

```python
y = data.iloc[:, 0]
```

```python
x = data.iloc[:, 1:]
```

```python
x
```

```python
y
```

## 모델 생성

### 모델

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(64,  activation='tanh'),   # hidden layer 1층 (노드 개수)
    tf.keras.layers.Dense(128, activation='tanh'),   # hidden layer 2층 (노드 개수)
    tf.keras.layers.Dense(1,   activation='sigmoid') # Output layer     (노드 1 개) **확률한개를 예측할 떄도 1개**
])
```

### 경하 and 손실 
- 경하 : adam, adagrad, adadelta, rmsprop, sgd 등
- loss : binary_crossentropy ( 결과가 0과 1상이의 분류/확률문제에서 씀 )

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
#                경하                    손실                     평가지표
```

## 학습
- model.fit(train, target, epochs) 
- epochs = 학습횟수

```python
model.fit(np.array(x), np.array(y), epochs=873)
```

## 예측

```python
pred = model.predict([ [750, 3.70, 3],
                       [400, 2.2, 1] ]) 
```
